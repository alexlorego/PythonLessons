**Задание 3**

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

> Для исследования влияния цены отформатируем датафрейм, оставив только те колонки, значения которых потенциально могут влиять на цену недвижимости.

```python
df.drop(['id','date', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1, inplace=True)
df
```

Построим корреляционную матрицу на основе имеющихся данных. По полученным коэфициентам мы сможем оценить силу связи между ценой и иными данными в таблице.


```python
corr_matrix = df.corr()
plt.figure(figsize=(20, 20))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm',fmt = '.1g')
```

> Согласно полученной матрице, наиболее выражена связь между ценой и жилой площадью(0.7) и ценой и оценкой дома(0.7). Чуть менее - цена и количество ванных(0.5). Связь вида и цены умеренная(0.4). Те с ростом перечисленных факторов цена на недвижимость пропорционально увеличивается.

Проведем графический анализ влияния конкретных характеристик на цену недвижимости.

Исследование 1: Цена и наличие вида на набережную

```python
data311 = df[df['waterfront'] == 0]['price']
data312 = df[df['waterfront'] == 1]['price']
plt.figure(figsize=(6, 4))
plt.boxplot([data311, data312], showfliers = False)
plt.title('Наличие вида на набережную')
plt.xticks([1, 2], ['Нет', 'Да'])
```

>На полученном графике видно, что наличие вида на набережную в среднем увеличивает стоимость недвижимости на 100тыс. Об этом можно судить по более высокому положению медианы и общему распределению цены недвижимости с видом на набережную. 

Исследование 2: Цена и наличие подвала

```python
data321 = df[df['sqft_basement'] == 0]['price']
data322 = df[df['sqft_basement'] > 0]['price']
plt.figure(figsize=(6, 4))
plt.boxplot([data321, data322], showfliers = False)
plt.title('Наличие подвала')
plt.xticks([1, 2], ['Нет', 'Да'])
```

> Согласно исследованию, наличие подвала также является значимым фактором при определении стоимости недвижимоcти. Среднестатистически объекты с подвалом стоят дороже, чем без него.

Исследование 3: Цена и состояние объекта

```python
sns.jointplot(x = df['price'], y = df['grade'], kind = 'reg')
```

> Наблюдается сильная прямая связь между ценой и оценкой недвижимости.

Исследование 4: Цена и год постройки

```python
sns.jointplot(x = df['price'], y = df['yr_built'], kind = 'reg')
```

> Можно заметить, что год постройки объекта недивижимости не является важным критерием при определении цены. На графике наблюдается отсутствие корреляции, об этом же можно сделать вывод по значению коэффициента корреляции = 0.05

Исследование 5: Цена и жилая площадь

```python
sns.jointplot(x = df['price'], y = df['sqft_living'], kind = 'reg')
```

> По нашим данным цена и жилая площадь обладают большой прямой линейной связью. С увеличением жилой площади объекта недвижимости, цена на него так же возрастает.
