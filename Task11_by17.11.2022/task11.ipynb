```python
import numpy as np
```

```python
from sympy import *
```

```python
import matplotlib.pyplot as plt
```

```python
xmin = -12
xmax = 12
count = 10000
```

```python
x = np.linspace(xmin, xmax, count)
```

```python
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
```

```python
График общей функции: (не понял из задания какой именно нужен, построил два)
```

```python
z1 = f + g
z2 = f - g
plt.plot(x, z1, label="f(x) + g(x)") 
plt.plot(x, z2, label="f(x) - g(x)")
plt.grid()
plt.legend()
plt.show()
```

```python
График фунцкий f(x) и g(x):
```

```python
plt.plot(x, f, label="f(x)")             
plt.plot(x, g, label="g(x)")             
plt.legend()
plt.grid()

idx = np.argwhere(np.diff(np.sign(f - g))).flatten() 
plt.plot(x[idx], f[idx], 'ro')
plt.show()
```

```python
print(f'Точек пересечения функций f(x) и g(x) -> {len(idx)} ')
```

```python
print(f'Функции пересекаются при x = {x[idx]}')
```

```python
Для проверки найдем точки пересечения иным способом, используя библиотеку sympy
```

```python
x = Symbol('x')
```

```python
fx = x**3 - 50*x
gx = -x**4 + 88*x**2 - 241
```

```python
inters = solve(fx - gx, x)
```

```python
inters_x = []
```

```python
for i in inters:
    z = round(i.n(10), 10)
    inters_x.append(z)
```

```python
print(f'Функции пересекаются при x = {inters_x}')
```
